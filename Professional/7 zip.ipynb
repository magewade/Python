```python
import zipfile

zip_file_name = 'workbook.zip'

file_count = 0

with zipfile.ZipFile(zip_file_name, 'r') as zip_file:
    for item in zip_file.infolist():
        if not item.is_dir():
            file_count += 1

print(file_count)
```

```python
import zipfile

zip_file_name = 'workbook.zip'

total_uncompressed_size = 0
total_compressed_size = 0

with zipfile.ZipFile(zip_file_name, 'r') as zip_file:
    for item in zip_file.infolist():
        if not item.is_dir():
            total_uncompressed_size += item.file_size
            total_compressed_size += item.compress_size

print("Объем исходных файлов: {} байт(а)".format(total_uncompressed_size))
print("Объем сжатых файлов: {} байт(а)".format(total_compressed_size))
```

```python
import zipfile
from datetime import datetime

zip_file_name = 'workbook.zip'

target_date = datetime(2021, 11, 30, 14, 22, 0)

matching_files = []

with zipfile.ZipFile(zip_file_name, 'r') as zip_file:
    for item in zip_file.infolist():
        if not item.is_dir():
            file_date = datetime(*item.date_time)
            
            if file_date > target_date:
                file_name = item.filename.split('/')[-1]
                matching_files.append(file_name)

matching_files.sort()

for file_name in matching_files:
    print(file_name)
```

```python
import zipfile
from datetime import datetime

zip_file_name = 'workbook.zip'

result = {} 

with zipfile.ZipFile(zip_file_name, 'r') as zip_file:
    for item in zip_file.infolist():
        if not item.is_dir():
            file_date = datetime(*item.date_time)
            
            file_name = item.filename.split('/')[-1]
            
            uncompressed_size = item.file_size
            
            compressed_size = item.compress_size

            result[file_name] = [file_date, uncompressed_size, compressed_size]

sorted_result = dict(sorted(result.items()))

for file_name, data in sorted_result.items():
    print(file_name)
    print(f"  Дата модификации файла: {data[0]}")
    print(f"  Объем исходного файла: {data[1]} байт(а)")
    print(f"  Объем сжатого файла: {data[2]} байт(а)")
    print()
```

```python
from zipfile import ZipFile

file_names = ['how to prove.pdf', 'fipi_demo_2022.pdf', 'Hollow Knight Silksong.exe',
              'code.jpeg', 'stepik.png', 'readme.txt', 'shopping_list.txt',
              'Alexandra Savior – Crying All the Time.mp3', 'homework.py', 'test.py']


with ZipFile('files.zip', 'w') as new_zip:
    for file_name in file_names:
        with open(file_name, 'rb') as file:
            file_data = file.read()
            if len(file_data) <= 100:
                new_zip.writestr(file_name, file_data)
```

```python
from zipfile import ZipFile

def extract_this(zip_name, *args):
    with ZipFile(zip_name, 'r') as zip_file:
        if not args:
            zip_file.extractall()
        else:
            for file_name in args:
                zip_file.extract(file_name)
```

```python
from zipfile import ZipFile

def convert_bytes(size, counter=0):
    sort_units = ('B', 'KB', 'MB', 'GB')
    while size > 1023:
        size /= 1024
        counter += 1
    return f' {round(size)} {sort_units[counter]}' if size else ''
    
with ZipFile('desktop.zip') as zip_file:
    for f in zip_file.infolist():
        name = f.filename.strip('/').split('/')
        spaces = (len(name) - 1) * '  '
        print(spaces + name[-1] + convert_bytes(f.file_size))
```
