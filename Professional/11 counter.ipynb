```python
from collections import Counter

def count_occurences(word, words):
    words = words.lower().split()
    word_counter = Counter(words)
    return word_counter[word.lower()]
    
    
word = 'Se'
words = 'se sdsf sds SE sdfsdg Se dhgf gfd asd se'

print(count_occurences(word, words))
```

```python
def calculate_group_cost(price_per_item, count):
    total_cost = price_per_item * count
    return total_cost

input_data = input().split(',')
product_dict = {}

for product in input_data:
    price_per_item = sum(ord(char) for char in product if char != ' ')
    
    if product in product_dict:
        product_dict[product]['count'] += 1
    else:
        product_dict[product] = {'price': price_per_item, 'count': 1}

sorted_products = sorted(product_dict.keys())

max_len = max(len(product) for product in sorted_products)

for product in sorted_products:
    price_per_item = product_dict[product]['price']
    count = product_dict[product]['count']
    total_cost = calculate_group_cost(price_per_item, count)
    print(f"{product: <{max_len}}: {price_per_item} UC x {count} = {total_cost} UC")


    
# рубашка,футболка,футболка,брюки,футболка,сырный соус,рубашка,носки,рубашка
# брюки      : 5425 UC x 1 = 5425 UC
# носки      : 5422 UC x 1 = 5422 UC
# рубашка    : 7574 UC x 3 = 22722 UC
# сырный соус: 10896 UC x 1 = 10896 UC
# футболка   : 8669 UC x 3 = 26007 UC
```

```python
from collections import Counter

with open('pythonzen.txt', 'r') as file:
    text = file.read()

cleaned_text = ''.join(char.lower() for char in text if char.isalpha())

letter_count = Counter(cleaned_text)

sorted_letters = sorted(letter_count.items())

for letter, count in sorted_letters:
    print(f"{letter}: {count}")
```

```python
from collections import Counter

words = Counter(input().lower().split())

result = []

for i in words.most_common():
    if i[1] == words.most_common()[-1][1]:
        result.append(i[0])
        
result.sort()

print(*result, sep=', ')
```

```python
import csv
from collections import Counter

changes_counter = Counter()

with open('name_log.csv', mode='r', encoding='utf-8') as file:
    reader = csv.reader(file)
    next(reader) 

    for row in reader:
        email = row[1]
        changes_counter[email] += 1

email_changes = Counter(changes_counter.values())

sorted_emails = sorted(changes_counter.keys())
for email in sorted_emails:
    changes = changes_counter[email]
    print(f"{email}: {changes_counter[email]}")

```
