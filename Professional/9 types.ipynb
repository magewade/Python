```python
alf = {chr(i):i-97 for i in range(97, 97+26)}
alphabet = list(input())
phrase = list(input().lower())
i = 0
for i in alf:
    alf[i] = alphabet[alf[i]]
    
for i in phrase:
    if i not in alf:
        print(i, end=' ')
    else:
        print(alf[i], end='')
```

```python
import csv
from collections import namedtuple
from datetime import datetime

Friend = namedtuple('Friend', ['surname', 'name', 'meeting_date'])

friends_data = []

def parse_datetime(date_str, time_str):
    date_format = "%d.%m.%Y %H:%M"
    return datetime.strptime(f"{date_str} {time_str}", date_format)

with open('meetings.csv', mode='r', encoding='utf-8') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)
    for row in csv_reader:
        meeting_date = parse_datetime(row[2], row[3])
        friend = Friend(row[0], row[1], meeting_date)
        friends_data.append(friend)

sorted_friends_data = sorted(friends_data, key=lambda friend: friend.meeting_date)

for friend in sorted_friends_data:
    print(f'{friend.surname} {friend.name}')
```
