```python
def is_valid(string):
    flag = False
    if (len(string) == 4 or len(string) == 5 or len(string) == 6) and string.isdigit():
        flag = True
    return flag

print(is_valid('92134'))
print(is_valid('89abc1'))
print(is_valid('49 83'))
```

```python
def print_given(*args, **kwargs):
    for i in args:
        print(i, type(i))
    for k, v in sorted(kwargs.items()):
        print(k, v, type(v))
        

print_given(1, [1, 2, 3], 'three', two=2)    
print_given(b=2, d=4, c=3, a=1)
```

```python
def convert(string):
    low_i = 0
    up_i = 0
    for i in list(string):
        if i.islower():
            low_i += 1
        if i.isupper():
            up_i += 1
    if low_i == up_i or low_i > up_i:
        return string.lower()
    else:
        return string.upper()
        

print(convert('BEEgeek'))
```

```python
def filter_anagrams(word, words):
    result = []
    for one_word in words:
        if sorted(word) == sorted(one_word):
            result.append(one_word)
    return(result)


word = 'abba'
anagrams = ['aabb', 'abcd', 'bbaa', 'dada']

print(filter_anagrams(word, anagrams))
```

```python
def likes(names):
    if len(names) == 0:
        return 'Никто не оценил данную запись'
    elif len(names) == 1:
        return f"{names[0]} оценил(а) данную запись"
    elif len(names) == 2:
        return f"{names[0]} и {names[1]} оценили данную запись"
    elif len(names) == 3:
        return f"{names[0]}, {names[1]} и {names[2]} оценили данную запись"
    else:
        return f"{names[0]}, {names[1]} и {len(names) - 2} других оценили данную запись"
    

print(likes(['Эндрю', 'Тоби', 'Том']))
```

```python
def index_of_nearest(numbers, number):
    difference = {i : abs(num - number) for i, num in enumerate(numbers)}
    return min(difference, key=difference.get)
    
print(index_of_nearest([9, 5, 3, 2, 11], 4))
```

```python
def spell(*args):
    result = {}
    for word in args:
        if word[0].lower() not in result:
            result[word[0].lower()] = len(word)
        elif (word[0].lower() in result.keys()) and (len(word) > result[word[0].lower()]):
            result[word[0].lower()] = len(word)
        else:
            continue
    return result

words = ['россия', 'Австрия', 'австралия', 'РумыниЯ', 'Украина', 'КИТай', 'УЗБЕКИСТАН']

print(spell(*words))
```

```python
def choose_plural(amount, declensions):
    if amount % 10 == 1 and amount % 100 != 11:
        plural_form = declensions[0]
    elif amount % 10 in [2, 3, 4] and amount % 100 not in [12, 13, 14]:
        plural_form = declensions[1]
    else:
        plural_form = declensions[2]
    result = f"{amount} {plural_form}"
    
    return result

print(choose_plural(21, ('пример', 'примера', 'примеров')))
```

```python
def get_biggest(numbers):
    if not numbers:
        return -1

    amount_dig_max_el = len(str(max(numbers))) 
    print(str(max(numbers)))
    numbers.sort(key=lambda x: str(x) * amount_dig_max_el, reverse=True)
    
    return int(''.join(map(str, numbers)))

print(get_biggest([61, 228, 9, 3, 11]))
```

```python
def sum_of_digits(number):
    return sum(map(int, str(number)))

n = int(input()) 

groups = {}

for number in range(1, n + 1):
    digit_sum = sum_of_digits(number)
    if digit_sum in groups:
        groups[digit_sum].append(number)
    else:
        groups[digit_sum] = [number]

max_group_len = max(len(group) for group in groups.values())

print(max_group_len)
```

```python
import math

# Создаем словарь для хранения данных о файловых группах
file_groups = {}

# Словарь для преобразования единиц измерения размера
size_file = {'B': 1, 'KB': 1024, 'MB': 1048576, 'GB': 1073741824}

# Открываем файл с явным указанием кодировки UTF-8
with open('files.txt', 'r', encoding='utf-8') as file:
    for line in file:
        # Разбиваем строку на части
        parts = line.split()
        if len(parts) == 3:
            filename, size_str, unit = parts
            # Извлекаем размер файла и преобразуем его в байты
            size_bytes = int(size_str) * size_file.get(unit, 1)
            file_extension = filename.split('.')[-1]
            
            # Добавляем файл в соответствующую группу
            if file_extension not in file_groups:
                file_groups[file_extension] = {'files': [], 'total_size': 0}
            
            file_groups[file_extension]['files'].append(filename)
            file_groups[file_extension]['total_size'] += size_bytes

# Сортируем группы файлов по расширению
sorted_groups = sorted(file_groups.items())

# Выводим результат
for extension, group_data in sorted_groups:
    print('\n'.join(sorted(group_data['files'])))
    total_size = group_data['total_size']
    summary = ""
    if total_size >= 1024 * 1024 * 1024:
        summary = f"Summary: {round(total_size / (1024 * 1024 * 1024))} GB"
    elif total_size >= 1024 * 1024:
        summary = f"Summary: {round(total_size / (1024 * 1024))} MB"
    elif total_size >= 1024:
        summary = f"Summary: {round(total_size / 1024)} KB"
    else:
        summary = f"Summary: {total_size} B"
    
    print('-' * 10)  # Добавляем разделительную полосу перед Summary
    print(summary)
    print()
```
