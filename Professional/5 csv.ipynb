```python
import pandas as pd 

name_log = pd.read_csv('/Users/magewade/Desktop/Python/Professional/name_log.csv')
name_log['dtime'] = pd.to_datetime(name_log['dtime'], format='%d/%m/%Y %H:%M')
name_log['dtime'] = name_log['dtime'].dt.strftime('%d/%m/%Y %H:%M')
name_log = name_log.sort_values(by='dtime', ascending=False)
name_log = name_log.drop_duplicates(subset=['email'], keep='first')
name_log = name_log.sort_values(by='email')
name_log.head()

name_log.to_csv('new_name_log.csv', encoding='utf-8', index=False)
```

```python
import pandas as pd

def condense_csv(filename, id_name):
    df = pd.read_csv('/Users/magewade/Desktop/Python/Professional/filename.csv', header=None, encoding='utf-8', dtype=str)
    pivoted = df.pivot(index=0, columns=1, values=2).reset_index()
    pivoted.columns.name = None
    pivoted.rename(columns={0: id_name}, inplace=True)
    pivoted = pivoted[[id_name] + list(df[1].unique())]
    pivoted.to_csv('condensed.csv', index=False, encoding='utf-8')

condense_csv('data.csv', id_name='ID')
```

```python
import pandas as pd
import numpy as np

prices = pd.read_csv('/Users/magewade/Desktop/Python/Professional/prices.csv', delimiter=';', index_col=0)
prices.head()
min_price = prices.iloc[:, 1:].values.min()
min_coords = [(row, col + 1) for row, col in zip(*np.where(prices.iloc[:, 1:].values == min_price))]
names = [(prices.columns[col], prices.index[row]) for row, col in min_coords]
sorted_names = sorted(names, key=lambda x: (x[0], x[1]))
shop, item = sorted_names[0][0], sorted_names[0][1]
print(f'{shop}: {item}')
```

```python
# Найдите минимальное значение
min_price = prices.iloc[:, 1:].min().min()

# Получите координаты минимального значения
min_coords = (prices == min_price).stack().idxmax()

# Получите название магазина и товара
shop, item = min_coords[1], min_coords[0]

print(f'{shop}: {item}')
```
